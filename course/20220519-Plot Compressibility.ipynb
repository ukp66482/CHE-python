```python
import numpy as np
import matplotlib.pyplot as plt

R = 0.082
T = 373

gas_name = ["N2","CH4","C2H6"]

A = [1.37, 2.3, 5.57]
B = [0.0387, 0.043, 0.06499]

for i in range(len(gas_name)):
    V = np.linspace(B[i]+0.03,1,100)
    P = R*T/(V-B[i])-A[i]/V**2
    Z=P*V/(R*T) 
    
    plt.plot(P, Z, label = gas_name[i])

plt.legend()
```
